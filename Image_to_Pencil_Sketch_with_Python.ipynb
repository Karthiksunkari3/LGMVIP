Interesting image filter effects, such as a pencil sketch or a cartoonizer effect, do not have to be very computationally involved to look good. In fact, in order to create a beautiful black-and-white pencil sketch effect, all you essentially need is some blurring and two image blending techniques called dodging and burning.



**Using OpenCV and Python, an RGB color image can be converted into a pencil sketch in four simple steps:**

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2 # importing opencv library packages
```

```python
image=cv2.imread("/content/Elon-musk-.jpg")
```

```python
cv2_imshow(image) # Displaying the original image
```

```python
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) # Original image ------>> Grayscale
```

**STEP 1 Convert the RGB color image to grayscale**

```python
cv2_imshow(img_gray)# Displaying the grayscale  image
```

**STEP 2 Invert the grayscale image to get a negative**

```python
img_invert=cv2.bitwise_not(img_gray) # Grayscale ------>> Negative
```

```python
cv2_imshow(img_invert) # Displaying the  Negative  image
```

**STEP 3 Apply a Gaussian blur to the negative from step 2**

```python
img_smooth=cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0) # Negative ------>> Blur image
```

```python
cv2_imshow(img_smooth) # Displaying the  blur image
```

**STEP 4 Blend the grayscale image from step 1 with the blurred negative from step 3 using a color dodge**

```python
def dodgeV2(x,y):
  return cv2.divide(x,255-y,scale=256)
```

```python
final_img=dodgeV2(img_gray,img_smooth) #Dividing the grayscale image by the inverted blurry image since images are just arrays
```

                            **FINAL PENCIL SKETCH IMAGE**

```python
cv2_imshow(final_img) # Displaying the  final sketch image
```
